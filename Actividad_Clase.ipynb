Generas 10,000 muestreos de una normal estándar, con media 0 y desviación estándar 1.

```python
# Importar librerias

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parametros
media_real = 0
desviacion_real = 1
N = 10000
```

```python
# Listas para almacenar las medias y desviaciones estándar

medias_muestrales = []
desviaciones_muestrales = []
```

# Luego de cada muestreo obtienes su media y su desviación estándar. Inicialmente no serán tan cercanos 0 y 1 respectivamente, pero se irán acercando conforme uses más observaciones para calcular estos estadísticos.


```python
# Generar muestreos y calcular estadísticos

for i in range(1, N + 1):
    muestra = np.random.normal(loc=media_real, scale=desviacion_real, size=i)
    media_muestral = np.mean(muestra)
    desviacion_muestral = np.std(muestra)

    # Guardar los valores
    medias_muestrales.append(media_muestral)
    desviaciones_muestrales.append(desviacion_muestral)
```

# Muestra en dos gráficas de líneas cómo van convergiendo esos resultados e interpreta lo que observas. En cada gráfica on una línea horizontal que indique dónde deberían converger estos valores: 0 para la gráfica que indica las medias observadas y 1 para la gráfica de las desviaciones observadas.


```python
# Crear las gráficas
plt.figure(figsize=(14, 6))

# Gráfica de las medias observadas
plt.subplot(1, 2, 1)
plt.plot(range(1, N + 1), medias_muestrales, label='Media muestral', color='r', alpha=0.7)
plt.axhline(y=media_real, color='r', linestyle='--', label='Media Teórica')
plt.title('Convergencia de la media muestral')
plt.xlabel('Número de muestras')
plt.ylabel('Media')
plt.legend()
```

```python
# Crear las gráficas
plt.figure(figsize=(14, 6))

# Gráfica de las desviaciones estándar observadas
plt.subplot(1, 2, 2)
plt.plot(range(1, N + 1), desviaciones_muestrales, label='Desviación estándar muestral', color='b', alpha=0.7)
plt.axhline(y=desviacion_real, color='r', linestyle='--', label='Desviación estándar teórica')
plt.title('Convergencia de la desviación estándar muestral')
plt.xlabel('Número de muestras')
plt.ylabel('Desviación estándar')
plt.legend()
```

# También responde ¿qué observas si aumentas la desviación estándar pero mantienes la media igual?

Al aumentar la desviación estándar, los datos estarán más dispersos alrededor de la media. Esto significa que habrá más variabilidad en los datos, y algunos valores estarán más alejados de la media que antes.

Si visualizas los datos,  la curva de distribución y los puntos de la gráfica se extienden más hacia los extremos en relación con la media.
